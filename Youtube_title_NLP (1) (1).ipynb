```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Dataset url
url='https://raw.githubusercontent.com/SAnand2903/BA830/main/Data/youtube.csv'
```

```python
# Reading the file
df=pd.read_csv(url)
```

```python
# Using regex to only keep English titled values in the dataset and saving it in english_text_df
import re
def is_english(text):
    # Regular expression to match only English text
    return bool(re.match(r'^[a-zA-Z\s.,!?-]+$', text))

# Filter the DataFrame to include only English text
english_text_df = df[df['title'].apply(is_english)]
```

```python
from nltk.corpus import stopwords
import nltk
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')

english_text_df["text_prcessed"] = english_text_df["title"].str.lower()
english_text_df["text_prcessed"] = english_text_df["title"].replace('<br />', ' ', regex=True)
english_text_df["text_prcessed"] = english_text_df["title"].replace('[^\w\d\s]', ' ', regex=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
cv.fit(english_text_df['text_prcessed'])

print('Number of tokens:', len(cv.vocabulary_))
print('Vocabulary:', cv.vocabulary_)
```

```python
my_stop_words = cv.get_stop_words()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_model = TfidfVectorizer(norm=None,stop_words=my_stop_words,max_features=500)
tfidf_model.fit(english_text_df['text_prcessed'])
df_tfidf_transformed = tfidf_model.transform(english_text_df['text_prcessed'])
tfidf_vectors = pd.DataFrame(df_tfidf_transformed.toarray(), columns=tfidf_model.get_feature_names_out(),dtype=np.float32)
tfidf_vectors
```

```python
tfidf_vectors = tfidf_vectors[0:5000]
```

Lets divide the data in train and test sets

```python
from sklearn.model_selection import train_test_split

X = tfidf_vectors
y = np.array(english_text_df["views"])[0:5000]
print(X.shape)
print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
y_train_scaled = scaler.fit_transform(y_train.reshape(-1, 1))
y_test_scaled = scaler.transform(y_test.reshape(-1, 1))

y_train_rav = y_train_scaled.ravel()
y_test_rav = y_test_scaled.ravel()
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
svr_model = SVR(kernel='linear')
```

```python
# Fit the SVR model on the scaled training data
svr_model.fit(X_train, y_train_rav)

# Predict on the training and testing data
y_train_pred = svr_model.predict(X_train)
y_test_pred = svr_model.predict(X_test)

# Evaluate the model
train_rmse = mean_squared_error(y_train_rav, y_train_pred, squared=False)
test_rmse = mean_squared_error(y_test_rav, y_test_pred, squared=False)

print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
```

```python
y_pred_train_inv = scaler.inverse_transform(np.array(y_train_pred).reshape(-1, 1))
y_pred_test_inv = scaler.inverse_transform(np.array(y_test_pred).reshape(-1, 1))
```

```python
train_rmse_inv = mean_squared_error(y_train, y_pred_train_inv, squared=False)
test_rmse_inv = mean_squared_error(y_test, y_pred_test_inv, squared=False)
```

```python
print("Train RMSE:", train_rmse_inv)
print("Test RMSE:", test_rmse_inv)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# Define a custom scoring function for RMSE
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Create TF-IDF vectorizer
tfidf_model = TfidfVectorizer(stop_words=my_stop_words)

# Create SVR model
svr_model = SVR()

# Create pipeline
pipeline = Pipeline([
    ('tfidf', tfidf_model),
    ('svr', svr_model)
])

y_train_scaled = scaler.fit_transform(y_train.reshape(-1, 1))
y_test_scaled = scaler.transform(y_test.reshape(-1, 1))

y_train_rav = y_train_scaled.ravel()
y_test_rav = y_test_scaled.ravel()

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(english_text_df['text_prcessed'][:5000],
                                                    np.array(english_text_df["views"][:5000]),
                                                    test_size=0.2,
                                                    random_state=42)

# Define hyperparameters grid for GridSearchCV
param_grid = {'tfidf__max_features': [500, 1000],
    'tfidf__norm': ['l1', 'l2'],
     'tfidf__min_df': [1, 2],
    'tfidf__max_df': [0.5, 0.75],
    'svr__kernel': ['linear', 'rbf'],
    'svr__C': [0.1, 1, 10],
    'svr__epsilon': [0.1, 0.2, 0.5]
}

# Define RMSE as the scoring metric for GridSearchCV
scoring = make_scorer(rmse, greater_is_better=False)

# Perform grid search
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring=scoring, n_jobs=-1)
grid_search.fit(X_train, y_train_rav)

# Best parameters and best score
print("Best parameters found:", grid_search.best_params_)
print("Best RMSE found:", np.abs(grid_search.best_score_))

# Predict on the testing data with the best model

y_train_pred = grid_search.predict(X_train)
y_test_pred = grid_search.predict(X_test)

#Taking inverse to calculate before scaled values
y_pred_train_inv = scaler.inverse_transform(np.array(y_train_pred).reshape(-1, 1))
y_pred_test_inv = scaler.inverse_transform(np.array(y_test_pred).reshape(-1, 1))


# Evaluate the best model
test_rmse = rmse(y_test, y_test_pred)
print("Test RMSE with best model:", test_rmse)

y_pred_train_inv = scaler.inverse_transform(np.array(y_train_pred).reshape(-1, 1))
y_pred_test_inv = scaler.inverse_transform(np.array(y_test_pred).reshape(-1, 1))

train_rmse_inv = mean_squared_error(y_train, y_pred_train_inv, squared=False)
test_rmse_inv = mean_squared_error(y_test, y_pred_test_inv, squared=False)

print("Train RMSE with inverse transform:", train_rmse_inv)
print("Test RMSE with inverse transform:", test_rmse_inv)
```

```python
from nltk.tokenize import word_tokenize

corpus = [t for t in english_text_df['text_prcessed'][0:5000].str.lower()]
```

```python
tokenised_corpus = [word_tokenize(t) for t in corpus]
```

```python
from gensim.models import Word2Vec
import numpy as np

# We construct and train our own Word2Vec
model_word2vec = Word2Vec(sentences=tokenised_corpus, vector_size=300, window=3, min_count=1, workers=4, negative=20, epochs=500)

print("All words captured by the model:", model_word2vec.wv.key_to_index)
# Get the embedding for each word captured by the model.
embeddings = np.array([model_word2vec.wv[word] for word in model_word2vec.wv.key_to_index])
```
